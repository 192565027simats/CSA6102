```python
import socket

# List of domains to investigate
domains = [
    "www.google.com",
    "www.python.org",
    "notarealdomain12345.com"
]

print("OSINT Domain Reconnaissance")
print("-" * 50)

for d in domains:
    try:
        ip = socket.gethostbyname(d)
        print(f"{d:30s} -> {ip}")
    except socket.gaierror:
        print(f"{d:30s} -> Could not resolve (invalid/unreachable)")
```
